```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

def warper(A,B,vis = False):    
    A = cv2.resize(A, (272, 192))
    B = cv2.resize(B, (272,192))
    
    G = A.copy()
    gpA = [G]
    for i in range(6):
        G = cv2.pyrDown(G)
        gpA.append(G)

    G = B.copy()
    gpB = [G]
    for i in range(6):
        G = cv2.pyrDown(G)
        gpB.append(G)
        
    lpA = [gpA[5]]
    for i in range(5,0,-1):
        GE = cv2.pyrUp(gpA[i])
        L = cv2.subtract(gpA[i-1],GE)
        lpA.append(L)

    lpB = [gpB[5]]
    for i in range(5,0,-1):
        GE = cv2.pyrUp(gpB[i])
        L = cv2.subtract(gpB[i-1],GE)
        lpB.append(L)  

    LS = []
    for la,lb in zip(lpA,lpB):
        rows,cols,dpt = la.shape
        #ls = cv2.addWeighted(la,0.1,lb,0.7,0)
        ls = np.hstack((la[:,0:int(cols/2)], lb[:,int(cols/2):]))
        LS.append(ls)
    
    if not vis:
        ls_ = LS[0]
        for i in range(1,6):
            ls_ = cv2.pyrUp(ls_)
            ls_ = cv2.add(ls_, LS[i])
            
    else:
        ls_ = LS[0]
        plt.figure(figsize=[18,13])
        plt.subplot(3,4,1);plt.imshow(ls_[:,:,::-1],aspect='auto');plt.title("Gaussian Merged Level {}".format(len(lpA) -1)); 

        def gennumbers():
            x = 1
            for i in range(10):
                x += 1
                yield x        
        y = gennumbers()

        for x in range(1,6):
            ls_ = cv2.pyrUp(ls_)
            plt.subplot(3,4,next(y));plt.imshow(ls_[:,:,::-1],aspect='auto');
            plt.title("Merged Pyramid PyrUp at level {}".format(len(lpA) - x)); 

            ls_ = cv2.add(ls_, LS[x])
            plt.subplot(3,4,next(y));plt.imshow(ls_[:,:,::-1],aspect='auto');
            plt.title("Laplacian injected at Level {}".format(len(lpA) -x));        
        
    return ls_    
```

```python
img1=cv2.imread('qa.jpg')
img2=cv2.imread('pm.jpg')
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')  

# perform the detection , here we are using 1.3 scale factor and 5 min neighbours
face1 = face_cascade.detectMultiScale(img1, 1.3, 5)
face2=face_cascade.detectMultiScale(img2, 1.3, 5)

print(face1)
for (x, y, w, h) in face1:
#     cv2.rectangle(img1, (x, y), (x+w, y+h), (0, 0, 255), 2)
    faceq = img1[y-40:y + h+40, x+10:x + w+40]
    #cv2.imshow("face",faces)
    cv2.imwrite('face.jpg', faceq)
for (x, y, w, h) in face2:
#     cv2.rectangle(img1, (x, y), (x+w, y+h), (0, 0, 255), 2)
    facei = img2[y-70:y + h+40, x-10:x + w+20]
    #cv2.imshow("face",faces)
    cv2.imwrite('face2.jpg', facei)
print(facei.shape)  
print(faceq.shape)   
plt.figure(figsize=[15,10])
plt.subplot(131);plt.imshow(faceq[:,:,::-1]);plt.title("Image 1");
plt.subplot(132);plt.imshow(facei[:,:,::-1]);plt.title("Image 2");
```

```python

combined = warper(faceq,facei)
plt.figure(figsize=[15,10])
plt.subplot(131);plt.imshow(combined[:,:,::-1]);plt.title("Image 1");
cv2.waitKey(0) # waits until a key is pressed
cv2.destroyAllWindows()

```
